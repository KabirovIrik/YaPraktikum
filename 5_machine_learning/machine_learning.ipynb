## 1. Откройте и изучите файл

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
data = pd.read_csv('/datasets/users_behavior.csv')
data.head(10)
```

## 2. Разбейте данные на выборки

Возьмем выборки следующим соотношением: <br>
Обучающая: **60%**<br>
Валидационная: **20%**<br>
Тестовая: **20%**<br>

```python
#Разбиваем датасет на обучающую и тестовую с валидационной выборками
df_train, df_validate_test = train_test_split(data, test_size=0.4, random_state=42)
# Получаем выборки для валидации и теста
df_validate, df_test = train_test_split(df_validate_test, test_size=0.50, random_state=42)

# Обучающая выборка
train_features = df_train.drop(['is_ultra'], axis=1)
train_target = df_train['is_ultra']

# Валидационная выборка
validate_features = df_validate.drop(['is_ultra'], axis=1)
validate_target = df_validate['is_ultra']

# Тестовая выборка
test_features = df_test.drop(['is_ultra'], axis=1)
test_target = df_test['is_ultra']

print('Размер Обучающей выборки', df_train.shape)
print('Размер Валидационной выборки', df_validate.shape)
print('Размер Тестовой выборки', df_test.shape)
```

## 3. Исследуйте модели

Так как задача - рекомендация тарифа, то будем использовать модели классификации: **LogisticRegression**, **RandomForestClassifier**, **DecisionTreeClassifier**. <br>

## 3.1 LogisticRegression

```python
model_lr = LogisticRegression(random_state=42)
model_lr.fit(train_features, train_target)
print('Оценка на обучающей выборке:', model_lr.score(train_features, train_target))
print('Оценка на валидационной выборке:', model_lr.score(validate_features, validate_target))
```

## 3.1 Вывод
Оценки не сильно отличаются, значит модель не переобучена. Но оценка меньше 0.75, так что эта модель не подойдет.

<hr>

## 3.2 RandomForestClassifier
Подберем 3 параметра: <br>
**n_estimators** <br>
**max_depth** <br>
**max_leaf_nodes** <br>

```python
best_model_rfc = None # Лучшая модель
best_score_rfc = 0 # Оценка лучшей модели 
score_rfc_list = [] # Список оценок на валидационой выборке
self_score_rfc_list = [] # Список оценок на обучающей выборке
best_x = 0
# Сделаем перебор циклами
for est in range(1, 15, 3):
    for depth in range(1, 20, 3):
        for leaf in range(2, 30, 3):
            # Создаем модель
            model_rfc = RandomForestClassifier(random_state=42, 
                                               n_estimators=est, 
                                               max_depth=depth, 
                                               max_leaf_nodes=leaf
                                              )
            
            model_rfc.fit(train_features, train_target) # Обучаем модель
            m_score = model_rfc.score(validate_features, validate_target) # Оценка модели на валидационой выборке
            train_score = model_rfc.score(train_features, train_target) # Оценка модели на обучающей выборке
            # Заносим оценки в списки
            score_rfc_list.append(m_score)
            self_score_rfc_list.append(train_score)
            # Если оценка текущей модели больше лучшей сохраненной, то сохраняем текущу.
            if m_score > best_score_rfc :
                best_score_rfc = m_score
                best_train_score = train_score
                best_model_rfc = model_rfc
                best_x = len(self_score_rfc_list)

print('Лучшая оценка на обучающей выборке:', best_train_score)
print('Лучшая оценка на валидационой выборке:', best_score_rfc)
print('Лучшая модель: \n', best_model_rfc)
```

<hr>

Получены следующие параметры: <br>
max_depth: 7 <br>
max_leaf_nodes: 14 <br>
n_estimators: 10 <br>

<hr>

Взглянем как подбиралась лучшая модель.

```python
plt.figure(figsize=(20,9))
plt.plot(score_rfc_list)
plt.plot(self_score_rfc_list)
plt.legend(['Валидационая', 'Обучающая'])
plt.title('Оценка на обучающей и валидационой выборках')
plt.annotate(
    'Лучшая модель', 
    xy=(best_x, best_score_rfc), 
    xytext=(best_x, 0.83), 
    arrowprops=dict(facecolor='black', shrink=0.05)
)
plt.show()
```

## 3.2 Вывод
Полуили модель с оценкой 0.81 на валидационой выборке. Разница с оценкой на обучающей выборке небольшая, значит переобучения нет.<br>


<hr>

## 3.3 DecisionTreeClassifier
Подберем 2 параметра: <br>
**max_leaf_nodes** <br>
**max_depth** <br>

```python
best_model_dtc = None
best_score_dtc = 0
score_dtc_list = [] 
self_score_dtc_list = [] 
best_x = 0
for depth in range(1, 30, 1):
    for leaf in range(2, 30, 3):
        model_dtc = DecisionTreeClassifier(random_state=42, max_depth=depth, max_leaf_nodes=leaf)
        model_dtc.fit(train_features, train_target)
        m_score = model_dtc.score(validate_features, validate_target)
        train_score = model_dtc.score(train_features, train_target)
        score_dtc_list.append(m_score)
        self_score_dtc_list.append(train_score)
        if m_score > best_score_dtc:
            best_score_dtc = m_score
            best_train_score = train_score
            best_model_dtc = model_dtc
            best_x = len(self_score_dtc_list)
print('Лучшая оценка на обучающей выборке:', best_train_score)
print('Лучшая оценка на валидационой выборке:', best_score_dtc)
print('Лучшая модель: \n', best_model_dtc)
```

<hr>

Получены следующие параметры: <br>
max_depth: 6 <br>
max_leaf_nodes: 8 <br>

<hr>

Взглянем как подбиралась лучшая модель.

```python
plt.figure(figsize=(20,9))
plt.plot(score_dtc_list)
plt.plot(self_score_dtc_list)
plt.legend(['Валидационая', 'Обучающая'])
plt.title('Оценка на обучающей и валидационой выборках')
plt.annotate(
    'Лучшая модель', 
    xy=(best_x, best_score_dtc), 
    xytext=(best_x, 0.83), 
    arrowprops=dict(facecolor='black', shrink=0.05)
)
plt.show()
```

## 3.3 Вывод
Данная модель точно почти такая же как и предыдущая, но точность хуже на 1%

## 4. Проверьте модель на тестовой выборке
Посмотрим оценки лучших моделей

```python
print('RandomForestClassifier: \t{:.3f}'.format(best_model_rfc.score(test_features, test_target)))
print('LogisticRegression: \t{:.3f}'.format(model_lr.score(test_features, test_target)))
print('DecisionTreeClassifier: \t{:.3f}'.format(best_model_dtc.score(test_features, test_target)))
```
