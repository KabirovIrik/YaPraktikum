Преобразовываем тип **first_day_exposition** в формат даты

```python
data['first_day_exposition_pr'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
data['first_day_exposition_pr'].head()
```

<hr>

Посмотрим и обработаем **ceiling_height**. У пентхаусов и больших помещений высота потолка может быть больше обычного, плюс они быват многоярусными.  

```python
data.sort_values('ceiling_height', ascending=False).head(20)
```

```python
data.hist('ceiling_height', bins=50, range=(5, 32))
```

```python
data.boxplot('ceiling_height')
```

Есть выброс в 100 метров. По гистограмме можно сказать, что от 20 до 32 метров есть много значений. Возможно, при их заполнении ошиблись с разрядностью, т.е. если поделить на 10, то эти данные станут вполне нормальными. Записей со значениями от 5 до 20 немного и маловероятно, что есть квартиры высотой до 2 метров, поэтому эти данные удалим.

```python
ceiling_height_median = data['ceiling_height'].median()
data['ceiling_height'] = data['ceiling_height'].fillna(ceiling_height_median)
```

```python
#пишем условие, обновлянем данные
cond = (data['ceiling_height'] > 20) & (data['ceiling_height'] <= 32)
data.loc[cond, 'ceiling_height'] = data.loc[cond, 'ceiling_height'] / 10
#избавляемся от лишнего
data = data.query('2.4 <= ceiling_height < 5')
data['ceiling_height'].describe()
```

<hr>

**floors_total** - пропуски нельзя заполнить средним, 0 или другими значениями. Можно было попробовать заполнить их значениями  столбца "floor", но это тоже не совсем корректно. Так как их мало, проще удалить. Также надо взглянуть на запись с 60 этажами.

```python
data[data['floors_total'].isnull()]
```

```python
data = data.query('floors_total.notnull()')
```

```python
data.sort_values('floors_total', ascending=False).head(10)
```

Согласно <a href="https://peterburg.center/ln/samye-vysokie-zdaniya-sankt-peterburga-top-10-dominant-v-severnoy-stolice-reyting-2019.html" target="_blank">этой статье</a>, нет жилых зданий в СПб в 60 и 52 этажей. Исключаем эти записи

```python
data = data.query('floors_total <= 36')
data['floors_total'].describe()
```

<hr>

Пропуски в **living_area** и **kitchen_area** вероятно значат, что помещения нежилые или забыли заполнить. Средним/медианным их заполнить не получится, т.к. могу быть несвопадения с общей площадью. Количество пропусков около 2000, что около 10%. Не рекомендуется удалять, лучше заполнить нулями.

```python
print('Пропуски kitchen_area', data['kitchen_area'].isnull().sum())
print('Пропуски living_area', data['living_area'].isnull().sum())
data['kitchen_area'] = data['kitchen_area'].fillna(0)
data['living_area'] = data['living_area'].fillna(0)
print('Пропуски kitchen_area', data['kitchen_area'].isnull().sum())
print('Пропуски living_area', data['living_area'].isnull().sum())
```

<hr>

Преобразуем и заполним **is_apartment**


```python
data['is_apartment'] = data['is_apartment'].fillna('False')
data['is_apartment'] = data['is_apartment'].map({'False':False, 'True':True})
data['is_apartment'] = data['is_apartment'].astype('bool')
print(data['is_apartment'].value_counts(), '\n')
data.info()
```

<hr>

Заполним пропуски **balcony** нулями

```python
print('Пропуски:', data['balcony'].isnull().sum())
data['balcony'] = data['balcony'].fillna(0)
print('Пропуски', data['balcony'].isnull().sum())
```

<hr>

**locality_name** - можно заполнить пропуски значением Санкт-Петербург, в тех строках, где указано расстояние до центра. Но сначала проверим: возможно это расстояния  указывают не только для СПб 

```python
data['locality_name'].isnull().sum()
```

```python
data[ (data['cityCenters_nearest'].notnull()) & (data['locality_name'] != 'Санкт-Петербург')]
```

Как видно, способ предложенный выше не подойдет. Тогда возьмем максимальное значение "cityCenters_nearest" в отфильтрованной таблице по городу СПб и заполним пропуски попавшие в этот диапазон, а остальные удалим.

```python
max_cityCenters_nearest = data[data['locality_name'] == 'Санкт-Петербург']['cityCenters_nearest'].max()
#Сохраним условие в переменной
cond = (data['cityCenters_nearest'] <= max_cityCenters_nearest) & (data['locality_name'].isnull())
data.loc[cond, 'locality_name'] = data.loc[cond, 'locality_name'].fillna('Санкт-Петербург')
print('Пропусков после заполнения', data['locality_name'].isnull().sum())
#Избавляемся от остатков
data = data.query('locality_name.notnull()')
print('Пропусков после удаления', data['locality_name'].isnull().sum())
```

<hr>

**parks_around3000**, **ponds_around3000** - можно заполнить нулями. Но сначала посмотрим на **parks_nearest** и **ponds_nearest** - если в пределах 3 км есть парк/водоем, то в parks_around3000/ponds_around3000 можно записать "1" <br>
И наоборот: если в parks_around3000/ponds_around3000 значения больше "0", то в parks_nearest/ponds_nearest можно записать "3000"

Сначала узнаем: есть ли записи подходящие под наши условия:

```python
print(len(data[( data['parks_around3000'].isnull()) & (data['parks_nearest'] <= 3000)]))
print(len(data[( data['parks_nearest'].isnull()) & (data['parks_around3000'] > 0)]))
print(len(data[( data['ponds_around3000'].isnull()) & (data['ponds_nearest'] <= 3000)]))
print(len(data[( data['ponds_nearest'].isnull()) & (data['ponds_around3000'] > 0)]))
```

Нужных нам записей нет.Тогда **_around3000** заполняем нулями, а в  **_nearest** оставим пропуски, т.к. любое числовое значение будет влиять на распределение данных.

```python
print('Пропуски:', data['parks_around3000'].isnull().sum())
data['parks_around3000'] = data['parks_around3000'].fillna(0)
print('Пропуски', data['parks_around3000'].isnull().sum())
```

```python
print('Пропуски:', data['ponds_around3000'].isnull().sum())
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)
print('Пропуски', data['ponds_around3000'].isnull().sum())
```

<hr>

Преобразуем **days_exposition** в целое

```python
cond = (data['days_exposition'].notnull())
data.loc[cond, 'days_exposition'] = data.loc[cond, 'days_exposition'].astype('int')
data['days_exposition'].describe()
```

Тип все равно остался float, но преобразование все равно необходимо было сделать, чтобы в дальнейшем можно было работать с данными *(при использовании метода .query() с некоторыми значениями запрос не срабатывал. Например, 90 или 45)*

**День недели, месяц и год публикации объявления**

<hr>

**Этаж квартиры**


```python
def set_name_floor(row):
    if row['floor'] == 1:
        return 'Первый'
    if row['floors_total'] == row['floor']:
        return 'Последний'
    return 'Другой'

def set_num_floor(row):
    if row['floor'] == 1:
        return 1
    if row['floors_total'] == row['floor']:
        return 2
    return 0

data['name_of_floor'] = data.apply(set_name_floor, axis=1)
data['num_of_floor'] = data.apply(set_num_floor, axis=1)
data[['num_of_floor', 'name_of_floor', 'floors_total', 'floor']].head(15)
```

<hr>

**Соотношение жилой и общей площади, а также отношение площади кухни к общей**

```python
data['part_kitchen_area'] = round((data['kitchen_area']/data['total_area'])*100, 2)
data['part_kitchen_area'].head()
```

### Шаг 4.3 Выбивающиеся значения

Для площади ограничимся значением до 300 кв.м. <br>

Количество дней до снятия с продажи: от 44 до 259, также исключим дни с выбросами

```python
cond1 = 'total_area <= 300 and '
cond2 = 'last_price <= 40000000 and '
cond3 = 'rub_per_m2 <= 300000 and '
cond4 = 'rooms >= 1 and '
cond5 = 'rooms <= 6 and '
cond6 = 'ceiling_height <= 3.5 and '
cond7 = 'days_exposition >= 44 and '
cond8 = 'days_exposition <= 259 and '
cond9 = 'days_exposition != 45 and '
cond10 = 'days_exposition != 90 and '
cond11 = 'days_exposition != 60 and '
cond12 = 'days_exposition != 30 and '
cond13 = 'days_exposition != 7'
good_data = data.query(cond1 + cond2 + cond3 + cond4 + cond5 + cond6 + cond7 + cond8 + cond9 + cond10 + cond11 + cond12 + cond13)
good_data.shape
```
